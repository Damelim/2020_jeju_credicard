```python
df =  pd.read_csv('C:/workspace/jeju/2020_jeju_creditcard/data/april_add_merge_with_homeproportion.csv',encoding = 'EUC-KR',index_col=0)
```

```python
df.isnull().sum() # no na values for total_sale_by_indst!!!
```

```python
iterative_columns = ['CARD_SIDO_NM','STD_CLSS_NM','HOM_SIDO_NM','AGE','SEX_CTGO_CD','FLC','year','month']

CARD_SIDO_NMs = df['CARD_SIDO_NM'].unique()
STD_CLSS_NMs  = df['STD_CLSS_NM'].unique()
HOM_SIDO_NMs  = df['HOM_SIDO_NM'].unique()
AGEs          = df['AGE'].unique()
SEX_CTGO_CDs  = df['SEX_CTGO_CD'].unique()
FLCs          = df['FLC'].unique()
years         = [2020]
months        = [7]

print(len(CARD_SIDO_NMs) , len(STD_CLSS_NMs),len(HOM_SIDO_NMs), len(AGEs), len(SEX_CTGO_CDs),len(FLCs)       )


temp = []
for CARD_SIDO_NM in CARD_SIDO_NMs:
    for STD_CLSS_NM in STD_CLSS_NMs:
        for HOM_SIDO_NM in HOM_SIDO_NMs:
            for AGE in AGEs:
                for SEX_CTGO_CD in SEX_CTGO_CDs:
                    for FLC in FLCs:
                        for year in years:
                            for month in months:
                                temp.append([CARD_SIDO_NM, STD_CLSS_NM, HOM_SIDO_NM, AGE, SEX_CTGO_CD, FLC, year, month])
temp = np.array(temp)
temp = pd.DataFrame(data=temp, columns = iterative_columns)
temp
```
